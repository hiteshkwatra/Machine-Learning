```python
import pandas as pd
import numpy as np
data=np.loadtxt("boston_train.csv",delimiter=",")
datatest=np.loadtxt("boston_test.csv",delimiter=",")
print(data.shape)
ltrain=data[:,:13]
print(ltrain.shape)
xtrain=np.insert(ltrain,len(ltrain[0]), values=1, axis=1)
print(xtrain.shape)
ytrain=data[:,13]
print(datatest)
df= pd.DataFrame(xtrain)
df
```

```python
import pandas as pd
df= pd.DataFrame(xtrain)
for i in range(len(xtrain[0])):
    for j in range(i,len(xtrain[0])):
        a=str(i*j)
        df[a]=df.iloc[:,i]* df.iloc[:,j]
df
xtrain=df.values
xtrain.shape
```

```python
def score(ypred,ytrue):
    a=((ytrue - ypred)**2).sum()
    b=((ytrue - ytrue.mean())**2).sum()
    return 1-(a/b)
```

```python
def gGD(m,a,inp,outp):
    mI_slope=np.zeros(len(inp[0]))
    l=len(inp)
    w = len(inp[0])
    for i in range(l):
        var = (-2/l)*(outp[i] - (m*inp[i]).sum())
        for j in range(w):
            mI_slope[j] += var*inp[i][j]
 
    mI_slope= m - a*(mI_slope)
    return  mI_slope
```

```python
def predict(m, test):
    l=len(test)
    y=np.ones(l)
    for i in range(l):
        y[i]=(m[:13]*test[i]).sum() + m[13]
    
    return y
```

```python
m=np.zeros(len(xtrain[0]))
iterations=1000
a=.1
for i in range(iterations):
    m=gGD(m,a,xtrain,ytrain)
```

```python

res=predict(m,ltrain)

print(res)

```

```python
la=score(res,ytrain)
print(la)
```
